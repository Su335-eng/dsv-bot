```python
# =========================
# Part 1: Upload and load fear_greed_index.csv
# =========================
import pandas as pd
from google.colab import files
import io
from IPython.display import display

print("Upload fear_greed_index.csv")
uploaded_sent = files.upload()  # Pick fear_greed_index.csv

sentiment_path = list(uploaded_sent.keys())[0]  # Get the uploaded filename
df_sent = pd.read_csv(io.BytesIO(uploaded_sent[sentiment_path]))

print("Sentiment data loaded:", df_sent.shape)
print("\nSentiment columns:", df_sent.columns.tolist())
print("\n--- Sample Sentiment Data ---")
display(df_sent.head())

# =========================
# Part 2: Upload and load historical_data.csv
# =========================
print("\n Upload historical_data.csv")
uploaded_trader = files.upload()  # Pick historical_data.csv

trader_path = list(uploaded_trader.keys())[0]  # Get the uploaded filename
df_trader = pd.read_csv(io.BytesIO(uploaded_trader[trader_path]))

print("Trader data loaded:", df_trader.shape)
print("\nTrader columns:", df_trader.columns.tolist()[:15])
print("\n--- Sample Trader Data ---")
display(df_trader.head())
```
